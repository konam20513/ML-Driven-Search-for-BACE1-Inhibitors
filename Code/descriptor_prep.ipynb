```python
import pandas as pd


# Read the TSV file into a DataFrame
df = pd.read_csv("data.tsv", sep='\t')
```

```python
df
```

```python
column_list = df.columns.tolist()
for i in column_list:
    print(i)
```

```python
# Count the number of NaN values in each row
df['NaN_Count'] = df.isnull().sum(axis=1)

# Sort the DataFrame by the 'Molecule ChEMBL ID' column and the number of NaN values
df_sorted = df.sort_values(by=['Molecule ChEMBL ID', 'NaN_Count'])

# Keep the rows with the least number of NaN values (first occurrence)
df_no_duplicates = df_sorted.drop_duplicates(subset='Molecule ChEMBL ID', keep='first')

# Remove the 'NaN_Count' column if not needed
df_no_duplicates = df_no_duplicates.drop(columns=['NaN_Count'])
```

```python
df_no_duplicates
```

```python
df=df_no_duplicates
```

```python
# Get the range of values in the 'Standard Value' column
min_value = df['Standard Value'].min()
max_value = df['Standard Value'].max()

# Display the range
print(f"Minimum Value: {min_value}")
print(f"Maximum Value: {max_value}")
```

```python
# Define a function to apply the labeling criteria
def label_ic50(value):
    if value < 1000:
        return 'active'
    else:
        return 'inactive'

# Add a new 'label' column based on the 'Standard Value' column
df['label'] = df['Standard Value'].apply(label_ic50)
```

```python
# Get the distribution of 'active' and 'inactive' values in the 'label' column
label_distribution = df['label'].value_counts()
label_distribution
```

```python
from mordred import Calculator, descriptors
from rdkit import Chem
import pandas as pd

# Create a Mordred calculator with all 2D descriptors
calc = Calculator(descriptors, ignore_3D=True)

# Assuming df is your DataFrame and it has a 'SMILES' column
molecules = [Chem.MolFromSmiles(smiles) for smiles in df['Smiles']]
desc_df = calc.pandas(molecules)
```

```python
import pandas as pd


# List of column names to select
selected_columns = ['nN', 'nO', 'AATS6dv', 'AATS0v', 'ATSC5d', 'GATS4Z', 'GATS5v', 'C1SP2', 'Xp-3dv', 'NtsC', 'SssssC', 'IC1', 'PEOE_VSA3', 'SMR_VSA4', 'n6ARing', 'GGI7']

# Create a new DataFrame from the selected columns
new_df = desc_df[selected_columns]

# Display the new DataFrame
print(new_df)
```
